Make a scatterplot of hwy vs. cty.



```r
ggplot() +
    geom_point(data = vehicles, 
               mapping = aes(x = cty, y = hwy))
```

Rename the axis titles to “Highway mpg” and “City mpg”, and add the chart title “Scatterplot of hwy vs. cty”.



```r
last_plot() + xlab("City mpg") + ylab("Highway mpg")
```

Modify the plot above such that the color of the dot represents cyl value.

```r
ggplot() +
    geom_point(data = vehicles, 
               mapping = aes(x = cty, y = hwy, col=cyl))+ xlab("City mpg") + ylab("Highway mpg")
```

```r
table(vehicles$cyl)
```

Modify the plot above so that each value of cyl is in its own plot.



```r
ggplot() +
    geom_text(data = vehicles, 
               mapping = aes(x = cty, y = hwy, col=cyl, label=cyl))+ xlab("City mpg") + ylab("Highway mpg")
```

```r
ggplot() +
    geom_point(data = vehicles, 
               mapping = aes(x = cty, y = hwy, col=factor(cyl)))+ xlab("City mpg") + ylab("Highway mpg")
```

Make a histogram of year.

```r
ggplot() + geom_histogram(data = vehicles, mapping = aes(year))
```

```r
vehicles %>% mutate( cyl = factor(cyl)) %>% filter(!is.na(cyl))
```
