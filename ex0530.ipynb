1. Make a scatterplot of hwy vs. cty.



````r
ggplot() +
    geom_point(data = vehicles, 
               mapping = aes(x = cty, y = hwy))
````

2. Rename the axis titles to “Highway mpg” and “City mpg”, and add the chart title “Scatterplot of hwy vs. cty”.



````r
last_plot() + xlab("City mpg") + ylab("Highway mpg") + ggtitle("Scatterplot of hwy vs. cty")
````

3. Modify the plot above such that the color of the dot represents cyl value.

````r
ggplot() +
    geom_point(data = vehicles, 
               mapping = aes(x = cty, y = hwy, col=cyl))+ xlab("City mpg") + ylab("Highway mpg")+
                 ggtitle("Scatterplot of hwy vs. cty")
````

````r
table(vehicles$cyl)
````

4. Modify the plot above so that each value of cyl is in its own plot.



````r
ggplot() +
    geom_text(data = vehicles, 
               mapping = aes(x = cty, y = hwy, col=cyl, label=cyl))+ xlab("City mpg") + ylab("Highway mpg")+
                 ggtitle("Scatterplot of hwy vs. cty")
````

````r
ggplot() +
    geom_point(data = vehicles, 
               mapping = aes(x = cty, y = hwy, col=factor(cyl)))+ xlab("City mpg") + ylab("Highway mpg")+
                 ggtitle("Scatterplot of hwy vs. cty")
````

5. Make a histogram of year.

````r
ggplot() + geom_histogram(data = vehicles, mapping = aes(year))
````

6. Convert the cyl column to an ordered factor variable, and remove all rows which have NA in the cyl column from the dataset.

7. For each value of cyl, make a violin plot of hwy values.



````r
vehicles %>% ggplot(aes(cyl, hwy, group=cyl)) + geom_violin()
````

9. Reduce the dataset to just rows with cyl equal to 4. Make a (jittered) scatterplot of hwy against year with alpha value 0.1. Add a geom_smooth layer without the SE bands. データセットを```cyl```が4と等しい行だけに減らせ． アルファ値0.1で，```year```に対する```hwy```の（jitteredな）散布図を作成せよ．SEバンドを除いた```geom_smooth```レイヤーを追加せよ．




````r
vehicles %>% filter(cyl==4) -> v4
v4
````

````r
v4 %>% ggplot() + geom_point(aes(year, hwy), alpha=0.1)
````

````r
v4 %>% ggplot(aes(year, hwy)) + geom_point(alpha=0.1) + geom_smooth( se = FALSE )
````

10. For the previous figure, change the shape of the points to crosses and change the theme to theme_minimal. 前の図で，点の形を十字に変え，テーマを```theme_minimal```に変更せよ．

````r
v4 %>% ggplot(aes(year, hwy)) + geom_point(alpha=0.1, shape="+") + geom_smooth( se = FALSE ) + theme_minimal()
````
